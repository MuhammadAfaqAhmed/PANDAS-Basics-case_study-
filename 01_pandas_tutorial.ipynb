# For multiple rows and columns

```python
#Within columns
df3.loc[:, ["X", "Z"]]
```

```python
#Within their respective rows and columns
# (:) this sign shows range
df3.loc["2022-01-01": "2022-01-11", ["W", "Z"]]
```

```python
#Within their respective rows and columns
# (,) this sign shows break e.g only the required values we want
df3.loc[["2022-01-01", "2022-01-11"], ["W", "Z"]]
```

```python
#Within their respective rows and columns
# for only 1 specifc value we need in respective row or column
df3.loc[["2022-01-01"], ["W", "X", "Y", "Z"]]
```

```python
#Specific value in specific column
#This is called scalear value
df3.at[dates[0], "X"]
```

```python
#Specific value in specific row
df3.iloc[3]
```

```python
#Specific value in specific row with specific range
df3.iloc[3:9]
```

```python
#Specific value in specific column
#0 is first column, 3 is third column
df3.iloc[:, 0:3]
```

```python
#Viewing the values in specifc column with specifc requirement
#in this we viewed only values greater than 0
#we can also view wih any range e.g >1, >3.5, > 2 etc
df3[df3["X"] > 0]
```

```python
#Viewing the values in complete data set
#in this we viewed only values greater than 0
#we can also view wih any range e.g >1, >3.5, > 2 etc
#NaN shows that it is empty or not correct data here
df3[df3>0]
```
